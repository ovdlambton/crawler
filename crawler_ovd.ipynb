```python
import requests
from bs4 import BeautifulSoup
```

BeautifulSoup is a popular Python Library used for web scraping. This is widely used by developers and data scientists to extract data from websites which can be used for data analysis, machine learning and data visualization.

```python
def crawl(url, depth=0, max_depth=5):
    if depth > max_depth:
        return

    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        print('  ' * depth + url)

        for link in soup.find_all('a'):
            href = link.get('href')
            if href.startswith('http'):
                crawl(href, depth + 1, max_depth)
    except:
        pass

crawl('https://support.spotify.com/us/article/payment-methods/')

```
